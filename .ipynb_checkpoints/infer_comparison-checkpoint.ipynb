## Inference

```python
from fastai.vision import *
from fastai.callbacks.hooks import *
from fastai.utils.mem import *
import os
from torchvision import transforms
from PIL import Image
import PIL
import matplotlib.pyplot as plt
import numpy as np
from torchvision import utils
import torch
import pandas as pd
```

```python
#pathfinding
path = os.getcwd()
img_path = path+'/data/h5'
path_img = img_path+'/img_wmasks'
path_lbl = path+'/data/h5/masks2/'
path_infer = img_path+'/s2-hr/'
path_infer_400 = img_path+'/s2-hr-400/'
path_infer_mask = img_path+'/s2-hr-mask/'
path_val = img_path+'/s2-lr-valid/'
path_naive = img_path+'/naive/'
il_val = ImageList.from_folder(path_val)
il_naive = ImageList.from_folder(path_naive)
```

```python
## misc code required to initialise learner object

fnames = get_image_files(path_img)
img_f = fnames[0]
lbl_names = get_image_files(path_lbl)
get_y_fn = lambda x: path_lbl+f'/{x.stem}_mask.tif'
mask = open_mask(get_y_fn(img_f))

src_size = np.array(mask.shape[1:])
codes = array(['Background' , 'Building'])
size = src_size//2


free = gpu_mem_get_free_no_cache()
# the max size of bs depends on the available GPU RAM
if free > 8200: bs=16
else:           bs=6
print(f"using bs={bs}, have {free}MB of GPU RAM free")

src = (SegmentationItemList.from_folder(path_img)
       .split_by_rand_pct()
       .label_from_func(get_y_fn, classes=codes))

data = (src.transform(get_transforms(), size=size, tfm_y=True)
        .databunch(bs=bs)
        .normalize(imagenet_stats))

name2id = {v:k for k,v in enumerate(codes)}
void_code = name2id['Background']

def acc_camvid(input, target):
    target = target.squeeze(1)
    mask = target != void_code
    return (input.argmax(dim=1)[mask]==target[mask]).float().mean()

metrics=acc_camvid
wd=1e-2

learn = unet_learner(data, models.resnet34, metrics=metrics, wd=wd)
```

```python
learn = load_learner(path_img)
```

```python
get_y_fn = lambda x: path_infer_mask+f'/{x.stem}.tif'
get_gtmask_fn = lambda x: path_lbl+f'/{x.stem}_mask.tif'
get_gtimg_fn = lambda x: path_img+f'/{x.stem}.png'
get_roc_fn = lambda x, suffix: img_path+'/roc/naive/'+f'{suffix}/{x.stem}.tif'
get_stem = lambda x: {x.stem}
```

```python
#create masks from interpolated images

t2im = transforms.ToPILImage()
results = []

for idx, im in enumerate(il_naive):
    pred = learn.predict(im)[2][1]
    L10 = (pred > 0.10).type(torch.uint8)
    t2im(L10).save(get_roc_fn(il_naive.items[idx], 'L10'))
    L15 = (pred > 0.15).type(torch.uint8)
    t2im(L15).save(get_roc_fn(il_naive.items[idx], 'L15'))
    L20 = (pred > 0.20).type(torch.uint8)
    t2im(L20).save(get_roc_fn(il_naive.items[idx], 'L20'))
    L25 = (pred > 0.25).type(torch.uint8)
    t2im(L25).save(get_roc_fn(il_naive.items[idx], 'L25'))
    L30 = (pred > 0.30).type(torch.uint8)
    t2im(L30).save(get_roc_fn(il_naive.items[idx], 'L30'))
    L35 = (pred > 0.35).type(torch.uint8)
    t2im(L35).save(get_roc_fn(il_naive.items[idx], 'L35'))
    L40 = (pred > 0.40).type(torch.uint8)
    t2im(L40).save(get_roc_fn(il_naive.items[idx], 'L40'))
    L45 = (pred > 0.45).type(torch.uint8)
    t2im(L45).save(get_roc_fn(il_naive.items[idx], 'L45'))
    L50 = (pred > 0.50).type(torch.uint8)
    t2im(L50).save(get_roc_fn(il_naive.items[idx], 'L50'))
```

```python
#true positive rate
def dice(pred, gt):
    num = 2*torch.logical_and(pred, gt).sum().item()
    denom = torch.logical_or(pred, gt).sum().item()+torch.logical_and(pred, gt).sum().item()
    return(num/denom)
    
def jaccard(pred, gt):
    num = torch.logical_and(pred, gt).sum().item()
    denom = torch.logical_or(pred, gt).sum().item()
    return(num/denom)
```

```python
##code for calculating ROC curves
get_naive_fn = lambda x, suffix: img_path+'/roc/naive/'+f'{suffix}/{x.stem}.tif'
get_gtmask_fn = lambda x: path_lbl+f'/{x.stem}_mask.tif'
get_hr_fn = lambda x, suffix: img_path+'/roc/'+f'{suffix}/{x.stem}.tif'
path_inferred_gt_mask = img_path+'/inferred_gt_masks/'
get_gtinfermask_fn = lambda x: path_inferred_gt_mask+f'/{x.stem}_mask.tif'


results = []
roc = "L30"

for idx, im in enumerate(il_val):
    gt_mask = (transforms.ToTensor()(Image.open(get_gtmask_fn(il_val.items[idx]))).squeeze()*255).bool()
    naive_mask = (transforms.ToTensor()(Image.open(get_naive_fn(il_val.items[idx], roc))).squeeze()*255).bool()
    hr_mask = (transforms.ToTensor()(Image.open(get_hr_fn(il_val.items[idx], roc))).squeeze()*255).bool()
    inferred_gt_mask = (transforms.ToTensor()(Image.open(get_hr_fn(il_val.items[idx], roc))).squeeze()*255).bool()

    dice_naive = dice(naive_mask, gt_mask)
    dice_hr = dice(hr_mask, gt_mask)
    jaccard_naive = jaccard(naive_mask, gt_mask)
    jaccard_hr = jaccard(hr_mask, gt_mask)

    results.append([set.pop(get_stem(il_val.items[idx])), dice_naive, dice_hr, jaccard_naive, jaccard_hr])
```

```python
res_pd = pd.DataFrame(results, columns =['Loc ID', 'Dice Bicubic', 'Dice ESRGAN', 'Jaccard Bicubic', 'Jaccard ESRGAN'])
```
