```python
import os
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import shutil
```

```python
path = os.getcwd()
img_path = path+'/data/h5'
path_lbl = path+'/data/h5/masks2/'
path_infer_400 = img_path+'/s2-hr-400/'
path_infer_mask = img_path+'/s2-hr-mask/'
path_gt_mask = img_path+'/s2-gt-mask/'
```

```python
hr_fn_list = []
hr_fn_list = sorted([f for f in os.listdir(path_infer_mask) if os.path.isfile(os.path.join(path_infer_mask, f))])
```

```python
gt_fn_source_list = []
for idx, line in enumerate(hr_fn_list):
    gt_fn_source_list.append(line.replace('.tif', '_mask.tif'))

for fn in gt_fn_source_list:
    shutil.copyfile(path_lbl+fn, path_gt_mask+fn)
```

```python
gt_fn_list = []
gt_fn_list = sorted([f for f in os.listdir(path_gt_mask) if os.path.isfile(os.path.join(path_gt_mask, f))])
```

```python
results = []

for gt, hr in zip(gt_fn_list, hr_fn_list):
    fn = hr.replace('.tif', '')
    gt_sum = np.array(Image.open(path_gt_mask+gt)).sum()
    hr_sum = np.array(Image.open(path_infer_mask+hr)).sum()
    results.append((fn, gt_sum, hr_sum))
```

```python
results_pd = pd.DataFrame(results, columns =['Loc_ID', 'Pix_GT', 'Pix_HR']) 
```

```python
results_pd.to_csv("./mask_results_exp1.csv", sep=',',index=False)
```
