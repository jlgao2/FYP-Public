```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
path = os.getcwd()
df = pd.read_csv(path+"/Training_log2.csv")
df
```

```python
plt.figure(figsize=(16,8))
plt.xlabel('Epochs')

ax1 = df["train_loss"].plot(color='blue', grid=True, label='Training Loss')
ax2 = df["valid_loss"].plot(color='red', grid=True, label='Validation Loss')

h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()

plt.axvspan(0, 9, color='red', alpha=0.1, label ="Stage 1")
plt.axvspan(9, 21, color='purple', alpha=0.1, label ="Stage 2")
plt.axvspan(21, 32, color='blue', alpha=0.1, label ="Stage 3")

plt.ylabel("Loss")

plt.legend(loc=3)
plt.show()
```
