```python
masks_path = path+'/data/h5/masks'
good_masks = path+'/data/h5/good_masks'
good_masks_hwy = path+'/data/h5/good_masks/hwy'
good_masks_bld = path+'/data/h5/good_masks/bld'
```

```python
img_path = path+'/data/h5/sem_seg'
new_img_path = img_path+'/new'
onlyfiles = [f for f in os.listdir(img_path) if os.path.isfile(os.path.join(img_path, f))]
```

```python
onlyfiles
```

```python
data = []
```

```python
for idx, line in enumerate(onlyfiles):
    data.append(line.replace('.png', ''))
    
```

```python
data
```

```python

results = []
def find(pattern, path):
    result = []
    for root, dirs, files in os.walk(path):
        for name in files:
            if fnmatch.fnmatch(name, pattern):
                result.append(os.path.join(root, name))
    return result

for line in data:
    results.append(find('*'+line+'*', masks_path))
```

```python
results
```

```python
for line in results:
    for subline in line:
        if 'highway' in subline:
            shutil.copy2(subline, good_masks_hwy)
        elif 'building' in subline:
            shutil.copy2(subline, good_masks_bld)
```
